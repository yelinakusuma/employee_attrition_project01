## **MODEL INFERENCE**

```python
# Import Libraries

import pandas as pd
import pickle
```

```python
# Memanggil model
with open('pipe_ada.pkl', 'rb') as model_file:
    pipe_ada = pickle.load(model_file)
```

**Data Employee 1**

```python
employee1 = {
    'Employee ID': 105000,
    'Age': 50,
    'Gender': 'Male',
    'Years at Company': 20,
    'Job Role': 'Technology',
    'Monthly Income': 8801,
    'Work-Life Balance': 'Good',
    'Job Satisfaction': 'High',
    'Performance Rating': 'High',
    'Number of Promotions': 2,
    'Overtime': 'Yes',
    'Distance from Home': 45.10,
    'Education Level': "Master's Degree",
    'Marital Status': 'Married',
    'Number of Dependents': 5,
    'Job Level': 'Senior',
    'Company Size': 'Large',
    'Company Tenure': 28,
    'Remote Work': 'No',
    'Leadership Opportunities': 'Yes',
    'Innovation Opportunities': 'Yes',
    'Company Reputation': 'Good',
    'Employee Recognition': 'Medium'
}

# Data dikonversi dalam DataFrame pandas
employee1 = pd.DataFrame([employee1])
employee1
```

```python
# Mengubah nama kolom
employee1.columns = employee1.columns.str.lower().str.replace(' ', '_')
employee1
```

```python
# Membuat fungsi transformasi label
def predicted(predict):
    if predict == 1:
        print("Stayed")
    else:
        print("Left")
```

```python
# Menjalankan model prediksi
predict1 = pipe_ada.predict(employee1)
employee1 = predicted(predict1)
```

Employee dengan id 105000 berhasil diprediksi dengan nilai 1, artinya karyawan tersebut diprediksi tidak akan resign di periode berikutnya.
